CONVOLUTION(3D):

```python
def convolution(data, filter, stride):
    result=np.zeros((int((data.shape[0]-filter.shape[0])/stride[0]) +1,int((data.shape[1]-filter.shape[1])/stride[1]) +1,int((data.shape[2]-filter.shape[2])/stride[2]) +1))
    for i in range(0, result.shape[0], 1):
        for j in range(0, result.shape[1], 1):
            for k in range(0, result.shape[2], 1):
                current_region=data[i*stride[0]:i*stride[0]+filter.shape[0],j*stride[1]:j*stride[1]+filter.shape[1],k*stride[2]:k*stride[2]+filter.shape[2]]
                result[i][j][k]=np.sum(current_region*filter)
    return result
```

DECONVOLUTION(3D):

```python
def deconvolution(data,filter,stride):
    result=np.zeros((int(data.shape[0]*stride[0]+filter.shape[0]-stride[0]),int(data.shape[1]*stride[1]+filter.shape[1]-stride[1]),int(data.shape[2]*stride[2]+filter.shape[2]-stride[2])))
    for i in range(0, data.shape[0], 1):
        for j in range(0, data.shape[1], 1):
            for k in range(0, data.shape[2], 1):
                current_region=data[i,j,k]*filter
                result[int(i*stride[0]):int(i*stride[0]+filter.shape[0]), int(j*stride[1]):int(j*stride[1]+filter.shape[1]), int(k*stride[2]):int(k*stride[2]+filter.shape[2])]=result[int(i*stride[0]):int(i*stride[0]+filter.shape[0]), int(j*stride[1]):int(j*stride[1]+filter.shape[1]), int(k*stride[2]):int(k*stride[2]+filter.shape[2])]+current_region
    return result
```

FORWARD PASS:

```python
def forward_pass(IP):
    a1=convolution(IP, conv_filter1, [1,2,2])
    a2=convolution(a1, conv_filter2, [1,1,1])
    a3=convolution(a2, conv_filter3, [1,1,1])    #bottleneck
    a4=deconvolution(a3, deconv_filter3, [1,1,1])
    a5=deconvolution(a4, deconv_filter2, [1,1,1])
    a6=deconvolution(a5, deconv_filter1, [1,2,2])
    return a6, a5, a4, a3, a2, a1
```

ERROR METRICS:

```python
def square_error(IP,OP):
    error=(IP-OP)**2
    #error=error*error
    #err=np.sum(error)/(IP.shape[0]*IP.shape[1]*IP.shape[2])
    return error
```

BACKWARD PASS:

```python
def backward_pass(IP, layers):
    a6, a5, a4, a3, a2, a1 = layers
    d_a6 = a6-IP  #3D

    d_deconv_filter1=convolution(d_a6, a5, [1,5,5])   #change
    #print(d_deconv_filter1.shape)
    d_a5=convolution(d_a6, np.transpose(deconv_filter1, (0,2,1)), [1,2,2])

    d_deconv_filter2=convolution(d_a5, a4, [1,1,1])
    #print(d_deconv_filter2.shape)
    d_a4=convolution(d_a5, np.transpose(deconv_filter2, (0,2,1)), [1,1,1])

    d_deconv_filter3=convolution(d_a4, a3, [1,1,1])
    #print(d_deconv_filter3.shape)
    d_a3=convolution(d_a4, np.transpose(deconv_filter3, (0,2,1)),[1,1,1])

    d_conv_filter3=convolution(a2,d_a3,[1,1,1])
    #print(d_conv_filter3.shape)
    d_a2=deconvolution(d_a3, np.transpose(conv_filter3, (0,2,1)), [1,1,1])

    d_conv_filter2=convolution(a1, d_a2, [1,1,1])
    #print(d_conv_filter2.shape)
    d_a1=deconvolution(d_a2, np.transpose(conv_filter2, (0,2,1)), [1,1,1])

    d_conv_filter1=convolution(IP, d_a1, [1,5,5])
    #print(d_conv_filter1.shape)

    return d_conv_filter1, d_conv_filter2, d_conv_filter3, d_deconv_filter1, d_deconv_filter2, d_deconv_filter3
```

```python
dataset_path="D:\PROJECT\CIFAR10\\train"

#Hypreparameters
epochs=15
learning_rate=0.0000000001
batch_size=1
image_size=[3,32,32]
```

```python
def load_image_paths():
    """Loads all image file paths from the dataset directory."""
    supported_formats = ('.png', '.jpg', '.jpeg', '.bmp')
    image_paths = [os.path.join(dataset_path, f) for f in os.listdir(dataset_path) 
                    if f.endswith(supported_formats)]
    return image_paths
    
def load_image(image_path):
    """Loads an image, resizes it, and normalizes pixel values to the range [0, 1]."""
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Could not load image: {image_path}")
    #Eimage = cv2.resize(image, image_size)
    image = image.astype(np.float64) / 255.0  # Normalize to [0, 1]
    return image
```
